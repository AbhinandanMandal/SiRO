**Visualizing Trends For 50 More Epochs**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

OOWL = pd.read_csv('/home/saswata/Downloads/SIOR-master/SingleEmbdTrainedResults/ObjectPI/1_OOWLbenchmark.csv')
MNet = pd.read_csv('/home/saswata/Downloads/SIOR-master/SingleEmbdTrainedResults/ModelNet40/1_MNet40benchmark.csv')
FG3D = pd.read_csv('/home/saswata/Downloads/SIOR-master/SingleEmbdTrainedResults/FG3D/1_FG3Dbenchmark.csv')



def extend_epochs(df, extra_epochs=50):
    last_epoch = df["epoch"].iloc[-1]
    new_epochs = np.arange(last_epoch + 1, last_epoch + 1 + extra_epochs)
    
    new_df = pd.DataFrame({"epoch": new_epochs})

    for col in df.columns:
        if col == "epoch":
            continue
        
        series = df[col]
        ema = series.ewm(span=5).mean()
        
        trend = ema.iloc[-1] - ema.iloc[-2]
        last_val = ema.iloc[-1]
        values = []

        for _ in range(extra_epochs):
            trend *= 0.9
            last_val = last_val + trend
            values.append(last_val)

        new_df[col] = values

    return pd.concat([df, new_df], ignore_index=True)


OOWL_ext = extend_epochs(OOWL)
MNet_ext = extend_epochs(MNet)
FG3D_ext = extend_epochs(FG3D)
```

*Train Loss vs Epoch*

```python

sns.set_theme(style="whitegrid")

plt.figure(figsize=(6,4))

plt.plot(OOWL_ext["epoch"], OOWL_ext["train_loss"], 
         label="ObjectPI", linewidth=1.5, linestyle="-")

plt.plot(MNet_ext["epoch"], MNet_ext["train_loss"], 
         label="ModelNet40", linewidth=1.5, linestyle="-")

plt.plot(FG3D_ext["epoch"], FG3D_ext["train_loss"], 
         label="FG3D", linewidth=1.5, linestyle="-")

plt.xlabel("Epoch")
plt.ylabel("Train Loss")
plt.title("Single Embedding Model +50 Epoch Extrapolation")

plt.legend(frameon=True)
plt.grid(True, linestyle="--")

sns.despine()
plt.tight_layout()

plt.savefig("TrainLoss_Epoch_SEM_50.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
def plot_with_split(df, label, linestyle):
    orig_len = len(df) - 50
    
    plt.plot(df["epoch"][:orig_len], df["train_loss"][:orig_len],
             linestyle=linestyle, linewidth=1.5, label=label)
    
    plt.plot(df["epoch"][orig_len:], df["train_loss"][orig_len:],
             linestyle=linestyle, linewidth=1.5)
    
plt.figure(figsize=(6,4))

plot_with_split(OOWL_ext, "ObjectPI", "-")
plot_with_split(MNet_ext, "ModelNet40", "-")
plot_with_split(FG3D_ext, "FG3D", "-")

plt.xlabel("Epoch")
plt.ylabel("Train Loss")
plt.title('Real vs Extrapolated')

plt.legend(frameon=True)
plt.show()
```
